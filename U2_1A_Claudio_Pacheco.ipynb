```python
#Crear funciones
im = np.array([[4, 5, 4, 4],
                          [8, 9, 10, 10],
                          [7, 8, 10, 10],
                          [8, 8, 12, 12]], dtype=np.uint8)
im
```

```python
def promedio_local(img, w,i,j):
    row_start = max(0, i - w)
    row_end = min(img.shape[0], i + w + 1)
    col_start = max(0, j - w)
    col_end = min(img.shape[1], j + w + 1)
    matriz = img[row_start:row_end, col_start:col_end]
    print('Se analiza: \n',matriz)
    print("El promedio local es: ",np.mean(matriz))
    
```

```python
promedio_local(im,1,1,1)
```

```python
def mediana_local(img, w, i, j):
    row_start = max(0, i - w)
    row_end = min(img.shape[0], i + w + 1)
    col_start = max(0, j - w)
    col_end = min(img.shape[1], j + w + 1)
    matriz = img[row_start:row_end, col_start:col_end]
    print('Se analiza: \n',matriz)
    print("La mediana local es: ",np.median(matriz))
```

```python
mediana_local(im,1,1,1)
```

### Anlisis de diferencias

Al realizar las operaciones con las herramientas de OpenCV se obtienen ciertos resultados. Por otra parte, al replicar estos pero utilizando la librera de Numpy, se obtienen resultados diferentes **en todos los casos**. Concretamente, se obtienen matrices ligeralmente diferentes. Algo que se aprecia es que cuando se realizan las operaciones con OpenCV, los resultados se truncan a 0 o 255.
