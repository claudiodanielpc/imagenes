```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def imagen(img):
    img=cv.imread('G:/.shortcut-targets-by-id/1Ow6KjVPo1myLESTKlM8Q0m1s8bsDHkGo/TratamientoDigitalImagenes_2022_2/Imagenes/'+img,0)
    if img is None:
        img=cv.imread('C:/Users/claudio.pacheco/Documents/Repositorio/'+img,0)
    plt.imshow(img,cmap='gray')
    return img
```

```python
img=imagen('DPP0357.TIF')
```

```python
#Incrementar la intensidad de los pixeles en 50
#Hacer una copia de img
img2 = img.copy()
#Incrementar la intensidad de los pixeles en 50
img2 = cv.add(img2,50)
plt.imshow(img2, cmap='gray')
```

```python
#Duplicar la intensidad de los pixeles
img3= img.copy()
img3 = img3*2
plt.imshow(img3, cmap='gray')
```

```python
#Suma y resta de dos imágenes distintas
imagen1=imagen('DPP0357.TIF')
imagen2=imagen('DPP0029.TIF')
```

```python
#Sumar las dos imagenes
suma = cv.add(imagen1,imagen2)
#Mostrar la imagen
plt.imshow(suma, cmap='gray')
```

```python
#Resta de las dos imagenes
resta = cv.subtract(imagen1,imagen2)
plt.imshow(resta, cmap='gray')
```
