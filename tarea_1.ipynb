```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
#Lectura de la imagen en escala de grises
img = cv.imread('G:/.shortcut-targets-by-id/1Ow6KjVPo1myLESTKlM8Q0m1s8bsDHkGo/TratamientoDigitalImagenes_2022_2/Imagenes/DPP0357.TIF',0)
plt.imshow(img, cmap='gray')
print("Dimensiones de la imagen: ", img.shape)
```

### a. Cambio de tamaño

```python
#Cambiar de tamaño. Doble
doble = cv.resize(img, None, fx=2, fy=2)
mitad = cv.resize(img, None, fx=0.5, fy=0.5)
```

```python
#Mostrar imágenes
titulos = ["Original " + str(img.shape), "Doble " + str(doble.shape), "Mitad " + str(mitad.shape)]
imagenes=[img,doble,mitad]

plt.figure(figsize=(15,5))
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(imagenes[i],cmap="gray")
    plt.title(titulos[i],fontsize=15,fontweight="bold")
    
```

### b. Recortar la imagen

```python
#Recortar la imagen a un tamaño de 512x512
crop_img= img[0:512, 0:512]
plt.imshow(crop_img, cmap='gray')
```

### b. Rotación de ángulos

```python
# Get the image's dimensions
height, width = img.shape[:2]
center = (width / 2, height / 2)
center
```

```python
# Create the rotation matrix
rotation_matrix = cv.getRotationMatrix2D(center, 5, 1)

# Rotate the image by 10 degrees
rotated_img = cv.warpAffine(img, rotation_matrix, (width, height))
plt.imshow(rotated_img)
```
